```python
from torch.utils.data import Dataset
import scipy
import os
from PIL import Image
from torch.utils.data import random_split
from torchvision import transforms
```

```python
# read a random image and try to convert it to tensor
img = '/Users/akshayranganath/Downloads/laya-katha-logo.jpg'
```

```python
img = Image.open(img)
```

```python
tensor = transforms.ToTensor()(img)
```

```python
tensor.shape
```

```python
tensor = transforms.Resize(200)(tensor)
```

```python
tensor = transforms.CenterCrop(256)(tensor)
```

```python
img = transforms.ToPILImage()(tensor)
```
